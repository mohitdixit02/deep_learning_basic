Error Functions - Neural Network Training

```python
import numpy as np

y_predicted = np.array([1, 1, 0, 0, 1])
y_true = np.array([0.3, 0.7, 1, 0, 0.5])
```

Mean Absoulute Error (MAE)

```python
def mean_absolute_error(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        total_error += abs(yt - yp)
    return total_error / len(y_true)
```

```python
mean_absolute_error(y_true, y_predicted).item()
```

```python
np.mean(np.abs(y_true - y_predicted)).item()
```

Log Loss (Logarithmic Loss)

```python
epsilon = 1e-15 # used to avoid log(0) in y_predicted_new
y_predicted_new = [max(i, epsilon) for i in y_predicted]
print(y_predicted_new)
y_predicted_new = [min(i, 1 - epsilon) for i in y_predicted_new]
print(y_predicted_new)
```

```python
np.log(y_predicted_new)
```

```python
-np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - np.array(y_predicted_new))).item()
```

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i, epsilon) for i in y_predicted]
    y_predicted_new = [min(i, 1 - epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new)).item()
    
```

```python
log_loss(y_true, y_predicted)
```

Mean Squared Error (MSE)

```python
def mean_squared_error(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        total_error += (yt - yp) ** 2
    return total_error / len(y_true)
```

```python
mean_squared_error(y_true, y_predicted).item()
```

```python
np.mean(np.square(y_true - y_predicted)).item()
```
